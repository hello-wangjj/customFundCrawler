```python
import requests
import re
import pandas as pd
```

```python
headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36 OPR/67.0.3575.79"}
    # urls = "http://fund.eastmoney.com/data/rankhandler.aspx?op=ph&dt=kf&ft=all&rs=&gs=0&sc=zzf&st=desc&sd=2019-03-19&ed=2020-03-19&qdii=&tabSubtype=,,,,,&pi=1&pn=5778&dx=1&v=0.4290949550138723"
urls = "http://fund.eastmoney.com/data/rankhandler.aspx"
    # sc=1nzf 近一年   sc=zzf 近一周  sc=1yzf 近一月  sc=3yzf 近3月  sc=6yzf 近6月  sc=2nzf 近2年  sc=3nzf 近三年 sc=jnzf 今年来  sc=lnzf 成立来
params = {"op": "ph", "dt": "kf", "ft": "all", "gs": "0", "sc": "1yzf", "st": "desc", "sd": "2019-03-24",
              "ed": "2020-03-24", "pi": "1", "pn": "6000", "dx": "1", "v": "0.4290949550138723"}
```

```python
r = requests.get(url=urls, headers=headers,params=params)
```

```python
r.url
```

```python
r.encoding
```

```python
text = r.text
```

```python
fundNum = re.compile(r'allNum:(\d*),',re.M)
fundNum = fundNum.findall(text)
fundNum
```

```python
pattern = re.compile(r'\d{6}')   # 查找基金代号
result1 = pattern.findall(text)
```

```python
len(result1)
```

```python
columns=['code']
save_file = pd.DataFrame(columns=columns, data=result1)
save_file.to_csv('fundList.csv', index=False, encoding="utf-8")
```

```python
pattern2 = re.compile(r'{.*}',re.M)
x = re.search(pattern2,text)
```

```python
x=x.group()
x
```

```python
fund_code = re.compile(r'(\d{6}),',re.M)
fc = re.findall(fund_code,x)
```

```python
len(fc)
```

```python
fund_name = re.compile(r'\d{6},(.*?),',re.M)
fn = re.findall(fund_name,x)
len(fn)
```

```python
fund_info = re.compile(r'"(\d{6},.*?)"',re.M)
fn = re.findall(fund_info,x)
len(fn)
```

```python
save_file = pd.DataFrame(data=fn)
save_file.to_csv('test.csv', index=False, encoding="utf-8")
```
